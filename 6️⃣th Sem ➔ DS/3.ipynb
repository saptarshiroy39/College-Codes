```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
columns = [
'class','Alcohol','Malic_acid','Ash','Alcalinity_of_ash','Magnesium',
'Total_phenols','Flavanoids','Nonflavanoid_phenols','Proanthocyanins',
'Color_intensity','Hue','OD280_OD315','Proline'
]

data = pd.read_csv("wine.data", names=columns)
X = data.drop('class', axis=1)
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

pca = PCA()

X_pca = pca.fit_transform(X_scaled)
print("Explained Variance Ratio of Each PC:")

for i, var in enumerate(pca.explained_variance_ratio_):
  print(f"PC{i+1}: {var:.4f}")
```

```python
X_plot = X_scaled[:, :2]
plt.figure(figsize=(10,6))
plt.scatter(X_plot[:,0], X_plot[:,1], alpha=0.6,color='gray', label='Data Points')
pc1 = pca.components_[0][:2]
pc2 = pca.components_[1][:2]
pc3 = pca.components_[2][:2]
plt.plot([0, pc1[0]*5], [0, pc1[1]*5], color='red', label='PC1')
plt.plot([0, pc2[0]*5], [0, pc2[1]*5], color='blue', label='PC2')
plt.plot([0, pc3[0]*5], [0, pc3[1]*5], color='green', label='PC3')
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Data Scatter with First 3 Principal Components")
plt.legend()
plt.show()
```
