```python
# From a list of integers, create a new list with only even numbers.

size = int(input("Enter Size of Array: "))
nums = []

for _ in range(size):
    nums.append(int(input()))

print(f"Original array: {nums}")

# nums_even = []
# for i in range(len(nums) - 1):
#     if nums[i] % 2 == 0:
#         nums_even.append(nums[i])
# print(nums_even)

nums_even = [i for i in nums if i % 2 == 0]
print(f"Even array elements: {nums_even}")
```

```python
# Write a Python program student_marks_analyzer.py that asks the user to enter the number of
# students in a class.Then ask the user to enter the marks (0–100) for each student, one by one,
# and store all the marks so that they can be processed later. After all marks are entered, the
# program should display:
# ● The list of all marks entered.
# ● The total of all marks.
# ● The average marks of the class.
# ● The highest and lowest marks in the class.
# ● How many students passed (marks ≥ 40).
# ● How many students failed (marks < 40).
# ● Display the marks in reverse order of entry.
# ● Display only those marks that are even numbers.


# student = int(input("Enter number of students: "))
# marks = []
# for _ in range(student):
#     marks.append(int(input("Enter mark: ")))

marks = list(map(int, input("Enter marks: ").split()))

print("Marks entered: ", marks)

# print("Total marks: ", sum(marks))
print(f"Total Marks: {sum(marks)}")

print(f"Average Marks: {sum(marks)/len(marks)}")

print(f"Highest Mark: {max(marks)}")

print(f"Lowest Mark: {min(marks)}")

pass_count = 0
for i in range(len(marks)):
    if marks[i] >= 40:
        pass_count += 1
print(f"{pass_count} students passed")

fail_count = 0
for i in range(len(marks)):
    if marks[i] < 40:
        fail_count += 1
print(f"{fail_count} students failed")

print(f"Marks in reversed order: {marks[::1]}")

even_marks = [i for i in marks if i % 2 == 0]
print("Even Marks:", even_marks)
```
